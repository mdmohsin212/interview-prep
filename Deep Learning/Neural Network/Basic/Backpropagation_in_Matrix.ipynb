```python
import numpy as np
```

```python
# Example: 2-layer network
# Input: (batch_size, input_dim)
# Weights: (input_dim, hidden_dim), (hidden_dim, output_dim)
def backpropagation_matrix_form():
    batch_size = 4
    input_dim = 3
    hidden_dim = 4
    output_dim = 2
    
    x = np.random.randn(batch_size, input_dim)
    y = np.random.randn(batch_size, output_dim)
    
    # Initialize weights
    w1 = np.random.randn(input_dim, hidden_dim) * 0.1
    b1 = np.zeros((1, hidden_dim))
    w2 = np.random.randn(hidden_dim, output_dim) * 0.1
    b2 = np.zeros((1, output_dim))
    learning_rate = 0.01
    
    # Forward pass
    Z1 = x @ w1 + b1
    A1 = 1 / (1 + np.exp(-np.clip(Z1, -250, 250)))
    Z2 = A1 @ w2 + b2
    A2 = 1 / (1 + np.exp(-np.clip(Z2, -250, 250)))
    
    # Loss (MSE)
    loss = np.mean((A2 - y) ** 2)
    
    # Backward pass
    m = batch_size
    
    # Output layer
    dA2 = 2 * (A2 - y) / m
    dZ2 = dA2 * A2 * (1 - A2)
    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)
    
    # Hidden layer
    dA1 = dZ2 @ w2.T
    dZ1 = dA1 * A1 * (1 - A1)
    dW1 = x.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)
    
    w1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    w2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    print(f"Loss: {loss:.4f}")
    print("Gradients computed and weights updated!")
    

backpropagation_matrix_form()
```
