```python
import emoji
import re
```

```python
def advanced_preprocessing(text):
    text = emoji.demojize(text)
    
    text = re.sub(r'@\w+', '<MENTION>', text)
    text = re.sub(r'#\w+', '<MENTION>', text)
    
    text = re.sub(r'\d+', '<NUMBER>', text)
    
    return text
```

```python
txt = "Just finished training my 1st model for 2024! 🔥 Shoutout to @AI_Guru for the tips on #MachineLearning. 100% ready for the next challenge! 🚀👍"

print(advanced_preprocessing(txt))
```
