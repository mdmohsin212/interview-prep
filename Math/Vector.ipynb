```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([1, 2, 3])
print(f"Vector: {v}")
print(f"Shape: {v.shape}")
print(f"Dimension: {v.ndim}")
```

```python
# Row vector
v_row = v.reshape(1, -1)
print(f'Row vector: {v_row}')
print(f"Row vector shape: {v_row.shape}") 
```

#### Vector Operations

```python
# Vector addition
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

v_sum = v1 + v2
print(f"Sum: {v_sum}")
```

```python
# Visual representation
fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1+v2')
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.legend()
plt.grid(True)
plt.show()
```

```python
# Scalar Multiplication
v = np.array([1, 2, 3])
scalar = 2

v_scaled = scalar * v
print(f"2 * v = {v_scaled}")
```

```python
# Dot Product (Inner Product)
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Formula: v1 · v2 = Σ(v1_i * v2_i) = 1*4 + 2*5 + 3*6 = 32
dot_product = np.dot(v1, v2)
dot_product2 = v1 @ v2

print(f"v1 · v2 = {dot_product, dot_product2}")
print(f"shape = {dot_product2.shape}")
```
