```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Rotation matrix (rotate 90 degrees)
R = np.array([[0, -1],
              [1,  0]])

# Original vector
v = np.array([1, 0])

# Transform
v_rotated = R @ v
print(f"Original: {v}")
print(f"Rotated: {v_rotated}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Before
ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
ax1.set_xlim([-1, 2])
ax1.set_ylim([-1, 2])
ax1.grid(True)
ax1.set_title('Before Rotation')

# After
ax2.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='b')
ax2.set_xlim([-1, 2])
ax2.set_ylim([-1, 2])
ax2.grid(True)
ax2.set_title('After Rotation')
plt.show()
```

```python
#Rotation Formula -> R = [[cos(θ) -sin(θ)],
#                         [sin(θ) cos(θ)]]
# For 70- degree rotation
theta = np.radians(70) # convert degree to radians
c, s = np.cos(theta), np.sin(theta)
R = np.array([[c, -s],
              [s, c]])

v_rotated_new = R @ v
print(f"Rotation Matrix for {70}°:\n{R}\n")
print(f"Rotated Vector: {v_rotated_new}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Before
ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
ax1.set_xlim([-1, 2])
ax1.set_ylim([-1, 2])
ax1.grid(True)
ax1.set_title('Before Rotation')

# After
ax2.quiver(0, 0, v_rotated_new[0], v_rotated_new[1], angles='xy', scale_units='xy', scale=1, color='b')
ax2.set_xlim([-1, 2])
ax2.set_ylim([-1, 2])
ax2.grid(True)
ax2.set_title('After Rotation')
plt.show()
```
