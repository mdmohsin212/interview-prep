#### Single Layer Neural Networks

**Forward pass in a neural network layer**
- output = activation(W @ input + b)

```python
input_size = 3
hidden_size = 4
output_size = 2

# weight metrices
w1 = np.random.rand(hidden_size, input_size)
b1 = np.random.rand(hidden_size)

w2 = np.random.rand(output_size, hidden_size)
b2 = np.random.rand(output_size)
```

```python
x = np.array([1, 2, 3])

# Forward pass
z1 = w1 @ x + b1
a1 = np.maximum(0, z1) # ReLU
z2 = w2 @ a1 + b2
output = z2

print(f"Input: {x}")
print(f"Output: {output}")
```

#### Linear Regression

**Normal equation ->** 
- w = (X^T @ X)^(-1) @ X^T @ y

```python
np.random.seed(42)
x = np.random.rand(100, 3)
true_weights = np.array([2, -1, 0.5])
y = x @ true_weights + 0.1 * np.random.rand(100)

weights = np.linalg.inv(x.T @ x) @ x.T @ y

print(f"True weights: {true_weights}")
print(f"Estimated weights: {weights}")
```
