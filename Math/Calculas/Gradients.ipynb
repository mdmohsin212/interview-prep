```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.differentiate import derivative
```

**Gradient**

- f(x, y) = x² + y²
- Gradient: ∇f = [2x, 2y]

```python
def f(x, y):
    return x**2 + y**2

def gradient(f, x, y, h=1e-5):
    df_dx = (f(x + h, y) - f(x - h, y)) / ( 2* h)
    df_dy = (f(x, y + h) - f(x, y - h)) / (2 * h)
    return np.array([df_dx, df_dy])

def gradient_exact(x, y):
    return np.array([2*x, 2*y])

x, y = 3, 4
grad = gradient_exact(x, y)
grad2 = gradient(f, x, y)
print(f"Gradient exact at (3,4): {grad}")
print(f"Gradient at (3,4): {grad2}")


# Gradient magnitude
grad_magnitude = np.linalg.norm(grad)
print(f"Gradient magnitude: {grad_magnitude:.2f}")
```

**Gradient Properties**

```python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Function values
Z = X**2 + Y**2

# Gradient components
U = 2*X  # ∂f/∂x
V = 2*Y  # ∂f/∂y

plt.contour(X, Y, Z, levels=20)
plt.quiver(X, Y, U, V, alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Field of f(x,y) = x² + y²')
plt.axis('equal')
plt.show()
```
